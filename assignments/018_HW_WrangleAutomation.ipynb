1. Write a function to convert the data file "018_HW_2x2_data.csv" from wide to "tidy" long format. Your function can assume that the input data consist of 4 columns (like our rat data), but should handle any number of rows (i.e. number of observations per cell).


```python
data1 = pd.read_csv('data/018_HW_2x2_data.csv')

def add_index_and_melt(data: pd.DataFrame) -> pd.DataFrame:
    '''
    Adds index column, melt and split column names into multiple columns

    Parameters
    ----------
    data : pd.DataFrame
        Data frame to be processed
        Data frame should be in wide format with column names in the format of A_1_B_2
    
    Returns
    -------
    pd.DataFrame
        Data frame in long format with index column
        Column names will be 'index', 'A', 'B', 'value'
    '''
    # save original index
    data = data.assign(index=data.index)
    # pivot wide to long
    data = data.melt(id_vars='index', value_vars=data.columns[:-1], var_name='variable', value_name='value')
    
    # split variable column into parts
    var_fields = data.variable.str.split('_')
    data.drop(columns='variable', inplace=True)

    # create new columns based on variable column (A_1_B_2 => A->1, B->2)
    var_columns = []
    for i in range(len(var_fields[0]) // 2):
        var_columns.append(var_fields.str[i*2][0])
        data[var_fields.str[i*2][0]] = var_fields.str[i*2 + 1]

    # reorder columns
    data = data[['index'] + var_columns + ['value']]

    return data

add_index_and_melt(data1)
```
