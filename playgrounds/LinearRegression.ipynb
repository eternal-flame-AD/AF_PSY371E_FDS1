```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.DataFrame({'x': np.arange(100, step=2), 'y': 2 * np.arange(100, step=2) + 1 + 20*np.random.randn(50)})
df
```

```python
lr = linregress(df.x, df.y)
```

```python
plt.plot(df.x, df.y, '.')
plt.plot(df.x, lr.intercept + lr.slope * df.x, 'r-')
```
