```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def run_beads(n_beads: int, n_steps: int):
    positions = np.zeros(n_beads)
    for _ in range(n_steps):
        positions += np.random.choice([-1, 1], size=n_beads)
    return positions

def plot_beads(positions):
    fig, ax = plt.subplots()
    ax.hist(positions, bins=20)
    ax.set_xlabel("Position")
    ax.set_ylabel("Count")

positions = run_beads(2000, 50)
plot_beads(positions)
```
